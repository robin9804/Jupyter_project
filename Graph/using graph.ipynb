```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
housing=pd.read_csv('/content/sample_data/california_housing_test.csv')
housing.head()
```

```
def graph(args1,args2):
  plt.plot(args1,args2,label="%str"%args2)
  plt.xlabel('wavelength')
  plt.ylabel('intensity')
  plt.show()
```

```
graph(housing['households'],housing['population'])
```

```
plt.scatter(housing['households'],housing['population'],label='population')
plt.legend()
```
