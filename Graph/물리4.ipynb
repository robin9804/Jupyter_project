```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from google.colab import drive
drive.mount('/gdrive')
```

```
rlc=pd.read_csv('/gdrive/My Drive/Colab Notebooks/data/RLC_RESULT.csv')
rlc
```

```
plt.plot(rlc['i(RC)'],rlc['Vc(RC)'],label='V-C')
plt.plot(rlc['i(RC)'],rlc['Vr(RC)'],label='V-R')
plt.plot(rlc['i(RC)'],rlc['Vrc(RC)'],label='V-RC')
plt.legend()
plt.grid()
```

```
def dit(fun):
  fp1=np.polyfit(rlc['i(RC)'],fun,1)
  f1=np.poly1d(fp1)
  print(f1)
dit(rlc['Vc(RC)'])
dit(rlc['Vr(RC)'])
dit(rlc['Vrc(RC)'])
```

```
plt.plot(rlc['i(RL)'],rlc['Vr(RL)'],label='V-R')
plt.plot(rlc['i(RL)'],rlc['Vl(RL)'],label='V-L')
plt.plot(rlc['i(RL)'],rlc['Vrl(RL)'],label='V-RL')
plt.legend()
plt.grid()
```
