##고찰
이번 실험은 코일에 교류 전원을 인가하여 1차 코일 안쪽에 위치한 2차 코일에 흐르는 교류 전압차이를 관찰하는 실험으로, 실험에 사용된 코일은 내부에 자기장을 발생시켜 2차 코일에 유도 전기장을 형성하여 교류 전위차를 측정할 수 있었다. 
실험의 결과를 통해 코일에서 만들어진 자기장이 유도한 전기장의 세기는 전류의 세기, 주파수, 전선을 감은 수, 2차 코일의 넓이에 비례하는 것을 알 수 있었다. 대부분의 변화에서 선형적인 비례를 확인할 수 있었으며, 이번 실험에서는 작은 범위의 전류만을 측정하여 부정확한 측정이 이루어졌기 때문에 더욱 정확한 전류 측정 기기가 있다면 정확한 경향성을 찾을 수 있을 것으로 보인다. 


```
d2={'V':[0.37,0.67,0.97],'turn':[25,30,40]}
turn2=pd.DataFrame(data=d2)
plt.plot(turn2['turn'],turn2['V'])
plt.title('Induction Coil Voltage Change of # of turn')
plt.xlabel('Radius of Coil [mm]')
plt.ylabel('Voltage [V]')
```
