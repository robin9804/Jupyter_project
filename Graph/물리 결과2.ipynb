```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
rst=pd.read_csv('/gdrive/My Drive/Colab Notebooks/data/Rst004.csv')
rst.head()
```

```
plt.plot(rst['A'],rst['f(a)_dyn'],label='data')
plt.xlabel('Current [A]')
plt.ylabel('Force [dyn]')
fp1=np.polyfit(rst['A'],rst['f(a)_dyn'],2)
f1=np.poly1d(fp1)
f1
plt.plot(rst['A'],f1(rst['A']),label='Polyfit')
plt.legend()
plt.grid()
f1
```

```
plt.plot(rst2['length'],rst2['f(l)'],label='data')
plt.xlabel('length [cm]')
plt.ylabel('Force [dyn]')
plt.legend()
fp2=np.polyfit(rst2['length'],rst2['f(l)'],1)
fp2
f2=np.poly1d(fp2)
plt.plot(rst2['length'],f2(rst2['length']),label='polyfit')
```

```
plt.plot(rst['n(magnt)'],rst['f(n)'],label='data')
plt.xlabel('# of magnet')
plt.ylabel('Force [dyn]')
plt.legend()
fp2=np.polyfit(rst['n(magnt)'],rst['f(n)'],1)
f2=np.poly1d(fp2)
plt.plot(rst['n(magnt)'],f1(rst['n(magnt)']),label='Polyfit')
f2
```
