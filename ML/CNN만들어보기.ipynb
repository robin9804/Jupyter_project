TF를 임포트하고 MNIST 다운받고있습니다.

```
(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()

train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

# 픽셀 값을 0~1 사이로 정규화합니다.
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```
# 합성곱 층을 만들기

model = models.Sequential()
# 이미지가 28x28 이미지임
model.add(layers.Conv2D(32, (3,3), activation='relu', input_shape=(28, 28, 1))) # 첫 층이라서 인풋 shape를 정해줌
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(64, (3,3), activation='relu')) # 출력 채널의 수
model.add(layers.MaxPool2D((2,2)))
model.add(layers.Conv2D(64, (3,3), activation='relu'))

model.add(layers.Flatten()) # 모델을 완성하기 위한 마지막 합성곱층의 출력크기 (3,3,64)를 하나의 Dense에 주입
model.add(layers.Dense(64, activation='relu')) # 출력을 1D로 받고, 3D로 출력하는 Dense함수
model.add(layers.Dense(10, activation='softmax')) # 소프트맥스 활성화
model.summary()
```

```
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(train_images, train_labels, epochs=5)
```

훈련중입니다. epoch는 훈련중에 사용하는 파라미터로, 총 몇 번의 횟수를 사용해 훈련을 시키는가를 나타내는 파라미터다

지금 예시로 사용한 epoch=5는 전체 데이터 셋에 대해서 전방, 후방으로 5번 왔다갔다한것을 의미한다.
